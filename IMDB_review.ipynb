```python
import os
import numpy as np
from keras.utils.np_utils import to_categorical
from keras.preprocessing import sequence
from keras.preprocessing.text import Tokenizer
```

```python
import re
def rm_tags(text):
    re_tag = re.compile(r'<[^>]+>')
    return re_tag.sub(' ', text)
```

```python
from io import open
def read_files(filetype):
    path = "aclImdb/"
    file_list = []
    
    positive_path = path + filetype + "/pos/"
    for f in os.listdir(positive_path):
        file_list += [positive_path + f]
        
    negative_path = path + filetype + "/neg/"
    for f in os.listdir(negative_path):
        file_list += [negative_path + f]
        
    print("read", filetype, "files:", len(file_list))
    
    all_labels = ([1] * 12500 + [0] * 12500)
    
    all_texts = []
    for fi in file_list:
        with open(fi, encoding = 'utf8') as file_input:
            all_texts += [rm_tags(" ".join(file_input.readlines()))]
            
    return all_labels, all_texts
```

```python
y_train, train_text = read_files("train")
```

```python
y_test, test_text = read_files("test")
```

```python
# build a ditionary that can contain 2000 words
token = Tokenizer(num_words=2000)

# count each word and insert to the dictionary
token.fit_on_texts(train_text)
```

```python
x_train_seq = token.texts_to_sequences(train_text)
x_test_seq = token.texts_to_sequences(test_text)
```

```python
# move all length in same length
x_train = sequence.pad_sequences(x_train_seq, maxlen=100)
x_test = sequence.pad_sequences(x_test_seq, maxlen=100)

# try to show the result of prune (if > 100)
print('before pad_sequences length=', len(x_train_seq[1]))
print(x_train_seq[1])
```

```python
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation, Flatten
from keras.layers.embeddings import Embedding
#build model of Embedding
model = Sequential()

model.add(Embedding(output_dim = 32,
                   input_dim = 2000,
                   input_length = 100))

model.add(Dropout(0.2))

model.add(Flatten())

model.add(Dense(units=256,
               activation = 'relu'))

model.add(Dropout(0.35))

model.add(Dense(units = 1,
               activation = 'sigmoid'))

model.summary()
```

```python
model.compile(loss = 'binary_crossentropy',
             optimizer = 'adam',
             metrics = ['accuracy'])

train_history = model.fit(x_train, y_train, batch_size=100, epochs=10, verbose=2, validation_split=0.2)
```

```python
scores = model.evaluate(x_test, y_test, verbose=1)
scores[1]
```
